```python
import os

def rename_files(directory='.'):
    # Get all files in the directory
    files = os.listdir(directory)
    
    # Filter for json files that match our pattern
    json_files = [f for f in files if f.endswith('_results.json')]
    
    # Counter for renamed files
    count = 0
    
    for old_name in json_files:
        # Create new name by replacing '_results.json' with '.json'
        new_name = old_name.replace('_results.json', '.json')
        
        # Construct full file paths
        old_path = os.path.join(directory, old_name)
        new_path = os.path.join(directory, new_name)
        
        try:
            # Rename the file
            os.rename(old_path, new_path)
            count += 1
            print(f"Renamed: {old_name} -> {new_name}")
        except Exception as e:
            print(f"Error renaming {old_name}: {str(e)}")
    
    print(f"\nSuccessfully renamed {count} files")

if __name__ == "__main__":
    # Simply use the current directory
    rename_files('.')

# If running in Jupyter/IPython, call the function directly
rename_files('.')
```
